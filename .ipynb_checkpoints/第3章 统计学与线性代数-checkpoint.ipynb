## 3.3 用Numpy进行线性代数运算

### 3.3.1 用Numpy求矩阵的逆

```python
A=np.mat('2 4 6;4 2 6;10 -4 18')
```

```python
A
```

```python
np.linalg.inv(A)
```

```python
A*np.linalg.inv(A)
```

```python
A*np.linalg.inv(A)-np.eye(3)
```

### 3.3.2 用Numpy解线性方程组

```python
A=np.mat('1 -2 1;0 2 -8;-4 5 9')
```

```python
b=np.array([0,8,-9])
```

```python
x=np.linalg.solve(A,b)
```

```python
x
```

```python
np.dot(A,x)
```

## 3.4 用Numpy计算特征值和特征向量

```python
A=np.mat('3 -2;1 0')
```

```python
evalue,evector=np.linalg.eig(A)
```

```python
evalue,evector
```

## 3.6 创建掩码式Numpy数组

```python
from scipy.misc import face
```

```python
face=face()[:,:,0]
random_mask=face>0
random_mask=random_mask.astype(int)
```

```python
random_mask=np.random.randint(0,2,size=face.shape)
```

```python
masked_array=np.ma.array(face,mask=random_mask)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.subplot(221)
plt.imshow(face)
plt.subplot(222)
plt.imshow(masked_array)
plt.subplot(223)
# plt.imshow(np.log(face))
plt.subplot(224)
plt.imshow(np.log(masked_array))
plt.show()
```
